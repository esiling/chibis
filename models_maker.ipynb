```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, matthews_corrcoef, confusion_matrix
from sklearn.metrics import make_scorer

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV, cross_val_score
import pickle


tox21 = pd.read_csv('melsums_total_norm.tsv', index_col=0, sep='\t')
unuque_file_names = tox21.index.tolist()
unuque_file_names_new = [str(x).split('.')[0].split('-')[0]+'_' +str(x).split('.')[0].split('-')[1] for x in unuque_file_names]
tox21['label'] = unuque_file_names_new
unuque_file_names_new = set(unuque_file_names_new)

for bird in unuque_file_names_new:
    a = tox21.loc[tox21['label'] == str(bird)]
    b = tox21.loc[tox21['label'] != str(bird)]
    a.drop('label', axis=1, inplace=True)
    b.drop('label', axis=1, inplace=True)
    a['label'] = 1
    b['label'] = 0
    c = pd.concat([a,b])
    y = c['label']
    
    X = c.drop('label', axis=1)
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    X = X.apply(lambda x: x.fillna(x.mean()),axis=0)
    
    model = KNeighborsClassifier()
    model.fit(X, y)
    
    with open('/home/veviorskiy/Tox21/birdhacker/ready_models/' + bird +'.pkl', 'wb') as f:
        pickle.dump(model, f)
    
```
